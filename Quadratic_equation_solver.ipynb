#### Quadratic equation solved using sympsy

This quadratic equation solver that will solve a second-order polynomial equation such as ax2 + bx + c = 0 for x, where a ≠ 0, using the quadratic formula.

The calculator solution will show work using the quadratic formula to solve the entered equation for real and complex roots. Calculator determines whether the discriminant (b2−4ac) is less than, greater than or equal to 0.

 When (b^2 − 4ac) = 0 there is one real root.

```python
one_real_root = sp.Eq((b**2)-4*a*c, 0)
one_real_root
```

 When (b^2 − 4ac)>0, there are two real roots.

```python
b**2-4*a*c > 0
```

When(b^2−4ac)<0, there are two complex roots

```python
b**2-4*a*c < 0
```

The above is used to solve quadratic equations where a ≠ 0 (polynomials with an order of 2)

```python
a,b,c,x = sp.symbols("a,b,c,x")
f = sp.Eq(a*x**2+b*x+c, 0)
f
```

The Discriminant are ± √((b^2) − 4ac)

1. Positive, there are 2 real solutions

2. Zero, there is one real solution

3. Negative, there are 2 complex solutions

### Case 1 (No real roots)

```python
%%time
def quadratic(a,b,c): 
#Quadratic formula for discriminant and zero points
    discriminant = (b**2) - (4*a*c) 
    dis_negative = (-b-(sp.sqrt(b**2-4*a*c)))/(2*a)
    dis_positive = (-b+(sp.sqrt(b**2-4*a*c)))/(2*a)
 #Control flow statement for finding root or no-root   
    if discriminant < 0: 
        print("\nThere is no real root")
    elif discriminant == 0:
        print("\nThere only one root")
    elif discriminant > 0:
        print("\nThere are two real roots")
#Control flow statement for finding parabola convex or concave 
    if a>0:
        print("\nParabola is convex")
    else:
        print("\nParabola is concave")
        
#Calculate parabola extreme coordinates      
#X-axis coordinate vertex
    x_coord_vertex = -b/(2*a)
#Y-axis coordinate vertex
    ax2 = (a*(-b/(2*a))**2)
    bx = (b*(-b/(2*a)))
    y_coord_vertex = ax2 + bx + c
#Sum root
    sum_root = (-b/a)
#Product root
    product_of_root = (c/a)

    return discriminant, dis_negative, dis_positive, x_coord_vertex, y_coord_vertex, sum_root, product_of_root

#Coefficients
print("INPUT:")
print("\nCoefficients")
a = float(input('Input coefficient a: '))  
b = float(input('Input coefficient b: '))  
c = float(input('Input coefficient c: ')) 
result = quadratic(a,b,c)

solution = [result]
discriminant_index = solution[0][0]
real_solution_index = solution[0][1]
complex_solution_index = solution[0][2]
x_coord_index = solution[0][3]
y_coord_index = solution[0][4]
sum_root = solution[0][5]
product_of_root = solution[0][6]
    
print("\nOUTPUT:")
print(f"Discriminant: {math.floor(discriminant_index)}")
print("\nPositive discriminant solution")
print(f"roots: {real_solution_index}")
print("\nNegative discriminant solution")
print(f"roots: {complex_solution_index}")
print(f"\nx-coordinate vertex: {x_coord_index}, y-coordinate vertex: {y_coord_index}")
print(f"\nSum root: {sum_root}")
print(f"\nProduct of root: {product_of_root}")
```
