```python
#Defining function
def algebra_equation(a,b):
    #algebric eqautions
    equation1 = (a-b)*(a+b)
    equation2 = (a**2)+(2*a*b)+(b**2)
    equation3 = ((a-b)**2) + (2*a*b)
    equation4 = (a**2) - (2*a*b) + (b**2)
    equation5 = (a**3) - (3*a**2*b)+(3*a*b**2) - (b**3)
    equation6 = (a+b)*(a**2 - a*b + b**2)
    equation7 = (a-b)*(a**2+a*b+b**2)
    #returning value
    return equation1, equation2, equation3, equation4, equation5, equation6, equation7
```

```python
algebra_equation(10, 20)
```
